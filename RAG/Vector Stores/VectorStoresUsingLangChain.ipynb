```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from langchain_core.documents import Document

doc1 = Document(
    page_content = "Virat Kohli is the face of RCB and one of the greatest batters in IPL history, known for his consistency and aggressive intent.",
    metadata= {"team":"Royal Challengers Bangalore"}
)
doc2 = Document(
    page_content = """MS Dhoni is the legendary captain of CSK, famous for his calm mindset and finishing abilities under pressure.
He transformed CSK into one of the most successful IPL teams and remains the biggest icon of the league.""",
    metadata= {"team":"Chennai Super Kings"}
)
doc3 = Document(
    page_content = """Rohit Sharma led Mumbai Indians to multiple IPL titles with his smart captaincy and elegant batting style.
Known as the “Hitman,” he is admired for scoring big runs effortlessly in crucial matches.""",
    metadata= {"team":"Mumbai Indians"}
)
doc4 = Document(
    page_content = """Andre Russell is KKR’s power-packed all-rounder who can change games within a few overs.
His explosive batting and wicket-taking ability make him one of the most dangerous T20 players.""",
    metadata= {"team":"Kolkata Night Riders"}
)
doc5 = Document(
    page_content = """Sanju Samson is the stylish captain of Rajasthan Royals, recognized for his elegant stroke play and fearless batting.
He has become the modern face of RR with his consistency and leadership growth.""",
    metadata= {"team":"Rajasthan Royals"}
)
```

```python
docs = [doc1,doc2,doc3,doc4,doc5]
```

```python
embedding = HuggingFaceEmbeddings(
    model="sentence-transformers/all-MiniLM-L6-v2"
)
vectorstore = Chroma(
    embedding_function= embedding,
    persist_directory='chroma-db',
    collection_name='sample'
)
```

```python
#add Documents to vector store
vectorstore.add_documents(docs)
```

```python
#get documents
vectorstore.get(include=['embeddings','documents','metadatas'])
```

```python
vectorstore.similarity_search(
    query="Who is best captain?",
    k=2
)
```
